```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
elements = ['Gender','Subscription Type','Contract Length']
for element in elements:
    df_customer[element] = encoder.fit_transform(df_customer[element])
```

```python
df_customer.info()
```

```python
df_customer.drop(columns=['CustomerID'],inplace=True)
```

```python
df_customer.isnull().sum()
```

```python
x = df_customer.drop(columns=['Churn'],axis=1)
y = df_customer["Churn"]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.3,random_state=43)
```

```python
print('Training input:', x_train.shape)
print('Training output:', y_train.shape)
print('Testing Input:',x_test.shape)
print('Testing output:',y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegressionCV

Model = LogisticRegressionCV()
Model.fit(x_train,y_train)
```
